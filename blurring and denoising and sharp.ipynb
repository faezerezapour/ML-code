```python
import cv2
import numpy as np
```

```python
image=cv2.imread("C:/Users/Raya-Center/Pictures/Saved Pictures/download.jpg")
image_blur=cv2.blur(image,(5,5))
cv2.imshow("my pic",image_blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image_blure=cv2.GaussianBlur(image,(5,5),0)
cv2.imshow("my pic",image_blure)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
bilateral=cv2.bilateralFilter(image,9,75,75)
cv2.imshow("my pic",bilateral)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
dst=cv2.fastNlMeansDenoisingColored(image,None,6,6,7,21)
cv2.imshow("my pic",dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

sharp

```python
kernel_sharpening=np.array([[-1,-1,-1],
                           [-1,9,-1],
                           [-1,-1,-1]])

sharpened=cv2.filter2D(image,-1,kernel_sharpening)

cv2.imshow("my pic",sharpened)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
