increasing Brightness

```python
import cv2
import numpy as np
```

```python
image=cv2.imread("C:/Users/Raya-Center/Pictures/Saved Pictures/download.jpg")
```

```python
M=np.ones(image.shape,dtype="uint8")*50
```

```python
print(M)
```

```python
added=cv2.add(image,M)
cv2.imshow("my pic",added)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

subtract

```python
M=np.ones(image.shape,dtype="uint8")*60
```

```python
subtracts=cv2.subtract(image,M)
cv2.imshow("my pic",subtracts)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
