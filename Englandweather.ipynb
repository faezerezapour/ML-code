```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data =pd.read_csv("F:/Dataset/EnglandWeather.csv") 
data
```

```python
df=pd.DataFrame(data) 
df
```

```python
df2=pd.DataFrame(df) 
df2
```

```python
df2[["Date","Time","Time Zone"]]=df2["Formatted Date"].str.split(" ",n=2,expand=True) 
df2
```

```python
date_str=pd.to_datetime(pd.Series(df2["Date"])) 
df2["Date"]=date_str.dt.strftime("%m/%d/%Y") 
df2 
```

```python
df3=pd.DataFrame(df2) 
df3
```

```python
new_date=df3["Formatted Date"].str.split(" ",n=2,expand=True) 
Year=pd.to_datetime(df3.Date).dt.year
df3["year"]=Year
Month=pd.to_datetime(df3.Date).dt.month
df3["month"]=Month
Hour=new_date[1].str.split(':',n=1,expand=True)
df3["hour"]=Hour[0] 
df3
```

```python
grouped=df3.sort_values(by="Summary").groupby("Precip Type") 
plt.scatter(grouped.get_group("rain")["Temperature (C)"],grouped.get_group("rain") 
["Humidity"]) 
plt.xlabel("Temperature (C) and rain",fontsize=10) 
plt.ylabel("Humidity",fontsize=10)
plt.savefig("C:/Users/Raya_Bit/Desktop/photo/0.jpg",bbox_inches="tight") 
plt.grid()
plt.show() 

```
